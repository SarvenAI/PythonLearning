```python
#import libraries
import pygame
import random
from pygame.locals import *

#initialise variables 
size = width,height =(800,800)
space_w = int(width/5)
startind = False
speed = 10
level = 1
counter = 1
white = (255,255,255)
score = 0

pygame.init()
run = True

#set window size
screen = pygame.display.set_mode(size)

#set text 
font = pygame.font.Font('freesansbold.ttf',22)
level_text = font.render(f'Level: {level}',True,white)
score_text = font.render(f'Scorw: {score}',True,white)

#set title
pygame.display.set_caption("Space game")

#load images 
spcaebg = pygame.image.load('Game Files/Space BG.png')
spacebg_loc = spcaebg.get_rect()

title = pygame.image.load('Game Files/Gametitle.png')
title_loc = title.get_rect()
title_loc.center = width/2,height/2

gamestart = pygame.image.load('Game Files/Start Msg.png')
gamestart_loc = gamestart.get_rect()
gamestart_loc.center = width/2,height/2

spceship = pygame.image.load('Game Files/Spaceship.png')
spceship_loc = spceship.get_rect()
spceship_loc.center = space_w,height*0.8

meteor = pygame.image.load('Game Files/Meteor.png')
meteor_loc = meteor.get_rect()
meteor_loc.center = space_w,height*0.1


gameover = pygame.image.load('Game Files/Game over.png')
gameover_loc = gameover.get_rect()
gameover_loc.center = width/2,height/2

#apply changes 
screen.blit(spcaebg,spacebg_loc)
screen.blit(gamestart,gamestart_loc)
screen.blit(spceship,spceship_loc)
screen.blit(meteor,meteor_loc)
screen.blit(title,title_loc)
screen.blit(level_text,(10,10))
screen.blit(score_text,(10,35))

pygame.display.update()

#start game sequence
while run:
    for event in pygame.event.get():
        if event.type == QUIT:
            run = False
        if event.type == KEYDOWN:
            if event.key == K_s:
                startind = True
            if event.key == K_r:
                startind = True
                level = 1
                score = 0
                counter = 1
                speed = 10
                meteor_loc.center = space_w,height*0.1
                spceship_loc.center = space_w,height*0.8
                level_text = font.render(f'Level: {level}',True,white)
                score_text = font.render(f'Scorw: {score}',True,white)
        while startind:
            counter = counter + 1
            if counter > 300:
                counter = 0
                level = level + 1
                speed = speed + level
                level_text = font.render(f'Level: {level}',True,white)
            for event in pygame.event.get():
                if event.type == KEYDOWN:
                    if event.key in [K_LEFT, K_a]:
                        spceship_loc = spceship_loc.move(-space_w,0)    
                    if event.key in [K_RIGHT, K_d]:
                        spceship_loc = spceship_loc.move(space_w,0)    
                if event.type == QUIT:
                    run = False
                    startind = False
            meteor_loc[1] = meteor_loc[1] + speed
            if meteor_loc[1] > height:
                meteor_loc.center = [space_w * random.randint(-1,5), -100]
                score = score + level
                score_text = font.render(f'Scorw: {score}',True,white)

            screen.blit(spcaebg,spacebg_loc)
            screen.blit(spceship,spceship_loc)
            screen.blit(meteor,meteor_loc)
            screen.blit(title,title_loc)
            screen.blit(level_text,(10,10))
            screen.blit(score_text,(10,35))
            
            #game over logic
            if spceship_loc[0] == meteor_loc[0] and spceship_loc[1] < meteor_loc[1] + 175 and spceship_loc[1]> meteor_loc[1] - 160:
                screen.blit(gameover,gameover_loc)
                pygame.display.update()
                startind = False
            pygame.display.update()
#quite game
pygame.quit()
```
